# Wine Quality Classification

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('wine.csv')
```

```python
df.head()
```

```python
#Turning strings in column "quality" into binary labels

df=df.replace("good",1)
df=df.replace("bad",0)
```

```python
#Here we see that fixed acidity does not give any specification to classify the quality.
fig = plt.figure(figsize = (5,5))
sns.barplot(x = 'quality', y = 'fixed acidity', data = df)
```

```python
#Here we see that its quite a downing trend in the volatile acidity as we go higher the quality 
fig = plt.figure(figsize = (5,5))
sns.barplot(x = 'quality', y = 'volatile acidity', data = df)
```

```python
#Composition of citric acid go higher as we go higher in the quality of the wine

fig=plt.figure(figsize=(5,5))
sns.barplot(x='quality',y='citric acid',data=df)
```

```python
# Correlation matrix
correlation_matrix = df.corr()

# Heatmap visualization
plt.figure(figsize=(15, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot for selected features
sns.pairplot(df[['alcohol', 'volatile acidity', 'sulphates', 'pH', 'quality']], hue='quality')
plt.show()
```

```python
#Let's start model training part

from sklearn.model_selection import train_test_split
X = df.drop(["quality"], axis = 1).values
y = df["quality"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

```python
#Applying Random forest algorithm
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(criterion = "gini", n_estimators = 90)
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

#### Analysis: 
The Random Forest model achieved an accuracy of 81% on the test set, indicating that the model correctly classified 81 out of every 100 wine samples based on their quality.

```python
from sklearn.metrics import classification_report

# Generate predictions
y_pred = model.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

#### Analysis:
Class-wise Performance:

- Class 0 (Low Quality):

1) Precision: 0.79  This means that 79% of wines predicted as low quality were actually low quality.
2) Recall: 0.78  This indicates that 78% of the actual low-quality wines were correctly identified by the model.
3) F1-score: 0.79  The balance between precision and recall is satisfactory, reflecting the model's reliability in classifying low-quality wines.

- Class 1 (High Quality):

1) Precision: 0.82  This indicates a higher confidence in the model's predictions for high-quality wines, with 82% accuracy in identifying them.
2) Recall: 0.82  The model successfully identified 82% of actual high-quality wines.
3) F1-score: 0.82  The strong F1-score suggests that the model performs well in both precision and recall for this class.
